```python
from google.colab import files
import pandas as pd

# Upload the XLSX file "niftyindex.xlsx" from your local PC
uploaded = files.upload()

# The 'uploaded' variable will contain the uploaded files
# You can access the file by its name, which in this case is "niftyindex.xlsx"
# Read the XLSX file into a Pandas DataFrame
file_name = "niftyindex.xlsx"  # Specify the filename
df = pd.read_excel(file_name)

# Display the first few rows of the DataFrame
df.head()
```

```python
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df' and contains 'Date' and 'Open' columns
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Open'], marker='o', linestyle='-', color='b', markersize=4)
plt.title('Open Price Over Time')
plt.xlabel('Date')
plt.ylabel('Open Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming your DataFrame is named 'df' and contains 'date', 'open', and 'low' columns
dates = df['Date']
open_prices = df['Open']
low_prices = df['Low']

# Define the width of the bars
bar_width = 0.35

# Create an array for the x-axis positions
x = np.arange(len(dates))

plt.figure(figsize=(12, 6))

# Create the bar plots for open and low prices
plt.bar(x, open_prices, width=bar_width, label='Open', color='blue')
plt.bar(x + bar_width, low_prices, width=bar_width, label='Low', color='green')

# Set the x-axis labels as dates
plt.xticks(x + bar_width / 2, dates, rotation=45)

plt.title('Open vs. Low Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
